```python
import os
import numpy as np
from face_utils import (
    add_student_embeddings, cosine_similarity,
    get_embedding, detector, mark_attendance_from_video,
    get_absent_students, mark_attendance_from_images
)
```

```python
# -----------------------------
# Embeddings file
# -----------------------------
student_embeddings_file = "student_embeddings.npz"
```

```python
# -----------------------------
# Add embeddings
# -----------------------------
student_name = "Ravi"
image_paths = ["prathmesh1.jpg"]

add_student_embeddings(student_name, image_paths, embeddings_file="student_embeddings.npz")
```

```python
# -----------------------------
# Load embeddings
# -----------------------------
student_embeddings = {}
if os.path.exists(student_embeddings_file):
    data = np.load(student_embeddings_file, allow_pickle=True)
    for key in data.keys():
        val = data[key].item() if hasattr(data[key], "item") else data[key]
        student_embeddings[key] = val
print(f"✅ Loaded embeddings for {len(student_embeddings)} students.")
```

```python
# -----------------------------
# Video Attendance
# -----------------------------
print("\n🎥 Starting video attendance session (press 'q' to quit)...")
attendance = mark_attendance_from_video(student_embeddings, video_source=0, threshold=0.7, duration=10)

print("\n📋 Attendance Log:")
for student, time in attendance.items():
    print(f"- {student} at {time}")
```

```python
# -----------------------------
# Image Attendance
# -----------------------------
test_folder = "test_stud"
print("\n🖼️ Starting image attendance check...")
attendance_from_images = mark_attendance_from_images(student_embeddings, test_folder, threshold=0.7)

print("\n📋 Attendance Log (Images):")
for student, time in attendance_from_images.items():
    print(f"- {student} at {time}")

absent_from_images = get_absent_students(student_embeddings, attendance_from_images)
print("\n❌ Absent Students (Images):")
for student in absent_from_images:
    print(f"- {student}")
```

```python
# -----------------------------
# Absent Students
# -----------------------------
absent = get_absent_students(student_embeddings, attendance)
print("\n❌ Absent Students:")
for student in absent:
    print(f"- {student}")
```
